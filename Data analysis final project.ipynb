# <div dir='rtl'> תובנה מס' 1:
 

 <div dir='rtl'> נסכום את סך כל המתים בכל קטגוריה בשנת 2019 על מנת לראות מה הסיבה שבגללה מתים הכי הרבה אנשים בעולם בשנה זו

```python
year2019= death_df.loc[(death_df.Year == 2019)].copy()
```

```python
df_new =year2019.sum(axis=0).reset_index()
df_new=df_new.rename(columns={'index': 'cause', 0:'total deaths in dozensd of millions'})
df_new.drop(df_new.index[[0,1 , -1]],inplace=True)
df_new.convert_dtypes()
df_new = df_new.sort_values(by=["total deaths in dozensd of millions"],ascending=False)
```

```python
fig4, ax = plt.subplots(figsize =(15,3))
sns.barplot(x='cause', y='total deaths in dozensd of millions', data=df_new)
sns.set_theme(style="darkgrid")
plt.xticks(rotation=90)
plt.ylabel('total deaths in dozensd of millions',size=20)
plt.xlabel('cause',size=20)
plt.title('Top Death Causes In The World- 2019',size=30, fontweight = 'bold')
plt.show()
```

 ###  <div dir='rtl'> ניתן לראות שרוב מקרי המוות בעולם בשנת 2019 נגרמו כתוצאה ממחלות לב וכלי דם - קרוב ל18.6 מיליון מקרים במקום השני נמצאת מחלת הסרטן עם קרוב 10 מיליון מקרים 

 <div dir='rtl'> נבדוק את אחוז החולים במחלות לב מתוך כל האוכלוסייה ונראה מי הן המדינות עם האחוז הגבוה ביותר ומי המדינה עם האחוז הנמוך ביותר

```python
perbyyear=year2019.groupby('Entity')[['Entity', 'Cardiovascular diseases','populaion']].max()
```

```python
per_pop = (perbyyear['Cardiovascular diseases'] / perbyyear['populaion'])*1000
perbyyear["Deaths per 1000 people"] = per_pop
```

```python
perbyyear.drop(per_pop.loc[perbyyear.populaion == 0].index, inplace = True)
```

```python
topDeaths= perbyyear.nlargest(3, 'Deaths per 1000 people')
topDeaths
```

###  <div dir='rtl'> ניתן לראות ששלוש המדינות עם אחוז המתים הגבוהה ביותר ממחלות לב ביחס לאוכולוסיה הן מדינות ממזרח אירופה, שהן ככל הנראה בעלות תקציב מוגבל יחסית בתחום הרפואה

```python
lowDeaths= perbyyear.nsmallest(1, 'Deaths per 1000 people')
lowDeaths
```

###  <div dir='rtl'> לעומת זאת בקטאר, המדינה העשירה בעולם נכון לשנת 2019, אחוז החולים במחלות לב הוא הנמוך ביותר ביחס לאוכולוסיה,  וככל הנראה קיימים האמצעים הכלכליים לטיפול במחלה זו 

```python
total2019 = perbyyear.loc[(perbyyear.Entity == 'Qatar') | (perbyyear.Entity == 'Bulgaria')]
```

 <div dir='rtl'> נבחן את ההבדל בין קטאר לבולגריה (שתי מדינות ה"קצה") - מתים ממחלות לב ביחס לאוכלוסייה

```python
sns.barplot(data = total2019, x='Entity', y='Deaths per 1000 people', color='Pink')
plt.show()
```

 <div dir='rtl'> ניתן לראות מהגרף שבשנת 2019 בבולגריה יש מעל 11 נפטרים ממחלות לב (פר אלף איש), ואילו בקטאר פחות מאדם אחד לאלף איש

###  <div dir='rtl'> כפי שהראנו קודם לכן סיבת המוות העיקרית בשנת 2019 היא מחלות לב, ככל הנראה הפערים בתמותה בין המדינות נובעים גם ממצב המדינה - מפותחת או מתפתחת- שכן למדינה מפותחת כמו קטאר יש את האמצעים לשפר את מצב התחלואה של האוכלוסיה בכל הקשור למחלות לב
